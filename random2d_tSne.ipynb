```python
import numpy as np
from sklearn.manifold import TSNE
import time
```

```python
df = np.random.rand(3000000,2)
```

```python
tsne = TSNE(n_components=2, random_state=42, n_jobs=25, verbose=10,n_iter=1500)
start = time.time()
X_tsne = tsne.fit_transform(df)
end = time.time()
print("Time taken by scikit learn:",(end-start))
with open("transformationData/random_2d_"+"scikit"+".npy", 'wb') as f:
    np.save(f,X_tsne)
    np.save(f,df)
```
