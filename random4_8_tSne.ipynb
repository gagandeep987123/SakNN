```python
import numpy as np
#import tSNE_reduce
from sklearn.manifold import TSNE
import time
```

```python
for i in range(4,9):
    print(str(i))
    df = np.random.rand(3000000,i)
    # start = time.time()
    # embedding = tSNE_reduce.openTSNE_transform(df)
    # end = time.time()
    # print("Time taken by opneTSNE:",(end-start))
    # with open("random_"+str(i)+"d_"+"opentsne"+".npy", 'wb') as f:
    #     np.save(f,embedding)
    #     np.save(f,df)

    tsne = TSNE(n_components=2, random_state=42, n_jobs=25, verbose=10,n_iter=1500)
    start = time.time()
    X_tsne = tsne.fit_transform(df)
    end = time.time()
    print("Time taken by scikit learn:",(end-start))
    with open("random_"+str(i)+"d_"+"scikit"+".npy", 'wb') as f:
        np.save(f,X_tsne)
        np.save(f,df)
    print("\n\n\n\n\n")
```
