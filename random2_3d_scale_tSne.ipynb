```python
import numpy as np
from sklearn.manifold import TSNE
import time
```

```python
data_lens = [10**5, 2* (10**5), 4 * (10**5), 8 * (10**5), 16*(10**5), 32*(10**5)]
```

```python
for dimension in [4]:
    for data_len in data_lens:
        data = np.random.rand(data_len,dimension)
        tsne = TSNE(n_components=2, random_state=42, n_jobs=25, verbose=10,n_iter=1500)
        start = time.time()
        X_tsne = tsne.fit_transform(data)
        end = time.time()
        print("Time taken by scikit learn:",(end-start))
        with open("scalingData/random_"+str(dimension)+"d_"+str(data_len)+"_scikit"+".npy", 'wb') as f:
            np.save(f,X_tsne)
            np.save(f,data)
```
